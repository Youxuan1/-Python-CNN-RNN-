如果将数据存储在 Python 列表或列表的嵌套列表中，则无法使用常规语法实现数据对位相加或相乘，但 ndarray 类可以实现

```python
import numpy as np

print("Python list operations:")
a = [1,2,3]
b = [4,5,6]
print("a+b:", a+b)
try:
    print(a*b)
except TypeError:
    print("a*b has no meaning for Python lists")
print()
print("numpy array operations:")
a = np.array([1,2,3])
b = np.array([4,5,6])
print("a+b:", a+b)
print("a*b:", a*b)
```

沿轴 0（二维数组的行）求和本质上就是沿该轴“折叠数组”，返回的数组比原始数组少一个维度

```python
a = np.array([[1, 2], [3, 4]])
print('a:')
print(a)
print('a.sum(axis=0)', a.sum(axis=0))
print('a.sum(axis=1)', a.sum(axis=1))
```

It is recommended to **comment the type** of variables in Python.

```python
import numpy as np

def square(x: np.ndarray) -> np.ndarray:
    ''' To calculate the square w.r.t. x
    :param x: x
    :return: x^2
    '''
    return np.power(x, 2)

def leaky_relu(x: np.ndarray) -> np.ndarray:
    ''' To calculate the leaky relu w.r.t. x
    :param x: x
    :return: max(eps*x, x)
    '''
    eps = 1e-3
    return np.maximum(eps*x, x)

x = np.arange(-5, 5, 0.1)
y1 = square(x)
y2 = leaky_relu(x)

import matplotlib.pyplot as plt
fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.plot(x, y1, color='tab:red', marker='o')
ax1.set_title('f(x)=x^2')
ax1.set_xlabel('x')
ax1.set_ylabel('square(x)')
ax1.grid(True)
ax2.plot(x, y2, color='tab:blue', marker='x')
ax2.set_title('f(x)=maximum(0, x)')
ax2.set_xlabel('x')
ax2.set_ylabel('leaky_relu(x)')
ax2.grid(True)
plt.show()
```
